# Step 0 

```python
import pandas as pd
import numpy as np
energy = pd.read_excel('Energy Indicators.xls', skiprows=16, skipfooter=38)
```

```python
energy.drop(energy.columns[[0, 1]], axis=1, inplace=True)

energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

energy.drop(energy.index[0], inplace=True)

energy.head()
```

```python
energy['Energy Supply'] = energy['Energy Supply']*1000000
```

```python
energy.replace(r'\..*\.', np.nan, inplace=True, regex=True)
```

```python
energy.head()
```

```python
energy.replace(to_replace=["Republic of Korea", "United States of America", "United Kingdom of Great Britain and Northern Ireland", "China, Hong Kong Special Administrative Region"], value=["South Korea", "United States", "United Kingdom", "Hong Kong"], inplace=True)
```

```python
energy['Country'] = energy['Country'].str.replace(r" \(.*\)","")
energy['Country'] = energy['Country'].str.replace(r"\d*","")
energy['Country'] = energy['Country'].str.strip()
```

```python
energy.head()
```

# Step 1

```python
GDP = pd.read_csv('world_bank.csv', skiprows=4)
GDP.head()
```

```python
GDP.replace(to_replace=["Korea, Rep.", "Iran, Islamic Rep.", "Hong Kong SAR, China"], value=["South Korea", "Iran", "Hong Kong"], inplace=True)
```

# Step 2

```python
ScimEn = pd.read_excel('scimagojr-3.xlsx')
ScimEn.head()
```

# Step 3

```python
GDP.columns
```

```python
GDP = GDP[['Country Name', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
GDP.rename(columns={'Country Name': 'Country'}, inplace=True)

print(GDP.shape)
print(ScimEn.shape)
print(energy.shape)

df = pd.merge(ScimEn, energy, how='outer', on='Country')
df = pd.merge(df, GDP, how='outer', on='Country')
df.set_index('Country', inplace=True)
```

```python
old = df.shape
print("Old", old)

df = df[df['Rank'] <= 15]

new = df.shape
print("New:", new)
df.sort_values(by='Rank', inplace=True)
df.head(15)
```

# Step 4

```python
print(abs(old[0]-new[0]), "lost")
```

# Step 5

```python
# Average GDP answer
avgGDP = df.loc[:, '2006':'2015'].mean(axis=1).sort_values(ascending=False)
avgGDP.head(15)
```

```python
meanEnergySupplyPerCapita = df['Energy Supply per Capita'].mean()
print(meanEnergySupplyPerCapita)
```

```python
maxRenewableName = df['% Renewable'].idxmax()
maxRenewablePercentage = df.loc[df['% Renewable'].idxmax(), '% Renewable']

print((maxRenewableName, maxRenewablePercentage))
```

# Step 6

```python
df['Self Citations / Total Citations'] = df['Self-citations'] / df['Citations']
maxRatioName = df['Self Citations / Total Citations'].idxmax()

print((maxRatioName, df.loc[maxRatioName, 'Self Citations / Total Citations']))
```

# Step 7

```python
df['HighRenew'] = df['% Renewable'].apply(lambda x: 1 if x >= df['% Renewable'].median() else 0)

df['HighRenew'].sort_values(ascending=False)
```
